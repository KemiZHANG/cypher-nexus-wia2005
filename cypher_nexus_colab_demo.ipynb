## Display Functions

````python
def flow_html(part_number):
    flow = ALGORITHM_FLOWS[part_number]
    nodes = []
    for index, step in enumerate(flow["steps"], start=1):
        nodes.append(f"""
        <div style='border:1px solid #4cc9c0;border-radius:8px;padding:12px;min-width:150px;flex:1;background:#081624;color:#eaf8f7;'>
            <div style='font-weight:800;color:#7ee7db;'>Step {index}</div>
            <div style='font-weight:800;margin:6px 0;'>{html.escape(step['title'])}</div>
            <div style='font-size:13px;line-height:1.35;color:#c8d8d9;'>{html.escape(step['detail'])}</div>
        </div>
        """)
    row = "<div style='font-weight:900;color:#ffcf5c;align-self:center;'>?</div>".join(nodes)
    note = html.escape(flow.get("note", ""))
    return f"""
    <div style='border:1px solid #2f7f7a;border-radius:10px;padding:14px;background:#06111d;margin:12px 0;'>
        <div style='display:flex;gap:8px;flex-wrap:wrap;align-items:stretch;'>{row}</div>
        <div style='margin-top:12px;border-left:4px solid #ffcf5c;padding:8px 10px;background:#18212a;color:#f7f7f7;'>
            <b>Flow explanation:</b> {note}
        </div>
    </div>
    """


def show_dataset_sample(part_number, rows=5):
    if part_number == 9:
        display(Markdown("**Dataset:** Derived from official outputs from Parts 1, 2, 3, and 8. No separate Excel sheet is used."))
        return
    df, meta = prepare_part_dataframe(part_number)
    dataset_file = meta.get("dataset_file") or meta.get("member") or Path(str(meta.get("source", ""))).name
    sheet_name = meta.get("sheet_name") or meta.get("sheet") or "N/A"
    row_count = meta.get("row_count", len(df))
    columns_used = ", ".join(meta.get("columns_used", list(df.columns)))
    display(Markdown(f"**Dataset:** `{dataset_file}` | **Sheet:** `{sheet_name}` | **Rows:** `{row_count}`  "))
    display(Markdown(f"**Columns used:** `{columns_used}`"))
    display(df.head(rows))


def show_code_for_part(part_number):
    for func in CODE_FUNCTIONS[part_number]:
        source = inspect.getsource(func)
        display(Markdown(f"### Code: `{func.__name__}`"))
        display(Markdown("```python\n" + source + "\n```"))


def show_output_for_part(part_number, max_chars=3500):
    result = ALL_RESULTS[f"part{part_number}"]
    output = result.get("output_text", "")
    if len(output) > max_chars:
        output = output[:max_chars] + "\n... output shortened for notebook display ..."
    print(output)


def show_part(part_number):
    info = PART_INFO[part_number]
    display(Markdown(f"# Part {part_number}: {info['title']}"))
    display(Markdown(f"**Selected algorithm:** {info['algorithm']}"))
    display(Markdown("**Candidate algorithms:** " + " / ".join(ALGORITHM_CHOICES[part_number]["options"])))
    display(Markdown("**Rejected algorithms:**\n" + "\n".join([f"- {reason}" for reason in info["rejected"]])))
    display(Markdown("## Dataset Evidence"))
    show_dataset_sample(part_number)
    display(Markdown("## Graphical Algorithm Explanation"))
    display(HTML(flow_html(part_number)))
    display(Markdown("## Implementation Code"))
    show_code_for_part(part_number)
    display(Markdown("## Official Output"))
    show_output_for_part(part_number)
````
